```python
#FINANCIAL ANALYSIS OF GLOBAL BANKS' STOCK PRICES
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
```

```python
BAC=data.DataReader('BAC','yahoo','2006')
```

```python
BAC
```

```python
BAC.head()
```

```python
#citigroup bank
C=data.DataReader('C','yahoo',start,end)
#Goldman Sachs
GS=data.DataReader('GS','yahoo',start,end)
#JP Morgan
JPM=data.DataReader('JPM','yahoo',start,end)
#Morgan Stanley
MS=data.DataReader('MS','yahoo',start,end)
```

```python
tickers=['BAC','C','GS','JPM','MS']
```

```python
#concatenation
bank_stocks=pd.concat([BAC,C,GS,JPM,MS],axis=1,keys=tickers)
bank_stocks.head()
```

```python
bank_stocks.columns
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks.head(2)
```

```python
bank_stocks['BAC']['Close'].max()
```

```python
for i in tickers:
    print("Bank=",i,";",bank_stocks[i]['Close'].max())
```

```python
#cross section
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
#to see how the returns are moving
#returns=(Value at time_t-value at time_(t-1))/(value at time_(t-1))
returns=pd.DataFrame()
```

```python
for i in tickers:
    returns[i+'return']=bank_stocks[i]['Close'].pct_change()
```
